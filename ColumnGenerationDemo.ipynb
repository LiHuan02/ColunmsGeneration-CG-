### ColumnGenerationSolver演示

本笔记本演示如何使用ColumnGenerationSolver类来解决下料问题并测试不同数据。

```python
# 导入ColumnGenerationSolver类
from CG import ColumnGenerationSolver
import numpy as np
```

```python
# 定义统一的求解器测试函数，接受三个输入参数
def run_solver_demo(L, lengths, demands, max_iter=50):
    """统一的下料问题求解器演示函数
    
    Args:
        title: 测试用例标题
        L: 原料长度
        lengths: 需求长度列表
        demands: 需求数量列表
        max_iter: 列生成最大迭代次数（默认50）
    """
    print(f"=== 原始问题数据 ===")
    print(f"原料长度: {L}")
    print(f"需求长度: {lengths}")
    print(f"需求数量: {demands}")
    
    solver = ColumnGenerationSolver(L, lengths, demands)
    try:
        patterns, sol, obj_relax = solver.column_generation(max_iter=max_iter)
        print(f"\n松弛问题目标值: {obj_relax:.6f}")
        print("生成的模式:")
        for i, p in enumerate(patterns):
            print(f"  模式 {i+1}: {p}")
        
        y_int, obj_int = solver.solve_integer_master(patterns)
        print(f"\n整数最优目标值: {obj_int}")
        print("各模式使用次数:")
        for i, count in enumerate(y_int):
            if count > 0:
                print(f"  模式 {i+1} 使用 {count} 次: {patterns[i]}")
    except Exception as e:
        print(f"求解过程中出现错误: {e}")
```

```python
L = 40
lengths = np.array([8, 12, 16, 20])
demands = np.array([90, 111, 55, 30])
run_solver_demo(L, lengths, demands)
```

```python
# 示例1: 原始问题数据
L = 20.0
lengths = [3, 7, 9, 16]
demands = [25, 30, 14, 8]

run_solver_demo(L, lengths, demands)
```
